```python
from backtesting import Backtest,Strategy
from backtesting.lib import crossover
```

```python
import yfinance as yf
import pandas as pd
import ta
```

```python
class SMAcross(Strategy):
    n1=50
    n2=100
    def init(self):
        close=self.data.Close
        self.sma1=self.I(ta.trend.sma_indicator,pd.Series(close),self.n1)
        self.sma2=self.I(ta.trend.sma_indicator,pd.Series(close),self.n2)
    def next(self):
            if crossover(self.sma1,self.sma2):
                self.buy()
            elif crossover(self.sma2,self.sma1):
                self.sell()
                
              
            
        
```

```python
df=yf.download('^NSEI',start='2018-01-01')
df
```

```python
bt=Backtest(df,SMAcross,cash=100000,commission=0.002,exclusive_orders=True)
```

```python
output=bt.run()
```

```python
output
```

```python
bt.plot()
```

```python
optim=bt.optimize(n1=range(50,160,10),n2=range(50,160,10),constraint=lambda x: x.n2-x.n1>20,maximize='Return [%]')
```
